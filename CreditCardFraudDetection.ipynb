After making these adjustments, you should have a 'Class' column with only binary values (0 or 1), where 0 represents non-fraudulent transactions, and 1 represents fraudulent transactions.

```python
credit_card_data['Class'] = credit_card_data['Class'].astype(int)
credit_card_data = credit_card_data[(credit_card_data['Class'] == 0) | (credit_card_data['Class'] == 1)]
```

TRY TO BALANCE DATASET:

*Isolation Forest is a powerful tool for anomaly (Outliers) detection *

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Separate the features (X) and target (y)
X = credit_card_data.drop('Class', axis=1)
y = credit_card_data['Class']

# Instantiate the Isolation Forest model
isolation_forest = IsolationForest(contamination='auto', random_state=42)

# Fit the model to the data
isolation_forest.fit(X)

# Predict outliers (fraudulent transactions)
y_pred = isolation_forest.predict(X)

# Convert predictions to binary labels (1 for inliers, -1 for outliers)
y_pred_labels = [1 if p == 1 else 0 for p in y_pred]

# Combine the predicted labels with the original data for further analysis
credit_card_data['Anomaly'] = y_pred_labels
```

```python

# Separate the data into legitimate (non-fraudulent) and potential fraudulent transactions
legit_transactions = credit_card_data[credit_card_data['Anomaly'] == 1]
potential_fraud_transactions = credit_card_data[credit_card_data['Anomaly'] == 0]
```

DATA AUGMENTATION:



```python
import pandas as pd
import numpy as np
from imblearn.under_sampling import RandomUnderSampler
# Separate the features (X) and target (y) for both legit and potential fraud transactions
X_legit = legit_transactions.drop('Class', axis=1)
y_legit = legit_transactions['Class']

X_fraud = potential_fraud_transactions.drop('Class', axis=1)
y_fraud = potential_fraud_transactions['Class']
desired_fraud_samples = 300
additional_fraud_samples_needed = desired_fraud_samples - len(X_fraud)

# Create additional samples through data augmentation (example: random perturbations)
augmented_fraud_samples = []
for _ in range(additional_fraud_samples_needed):
    random_idx = np.random.randint(0, len(X_fraud))
    new_sample = X_fraud.iloc[random_idx].copy()
    new_sample += np.random.normal(0, 0.01, len(new_sample))
    augmented_fraud_samples.append(new_sample)

# Convert the augmented samples list to a DataFrame
augmented_fraud_transactions = pd.DataFrame(augmented_fraud_samples, columns=X_fraud.columns)
```
